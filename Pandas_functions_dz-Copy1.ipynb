Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
dfrating = pd.read_csv('ratings.csv')
dfrating.head()
```

```python
df_low = dfrating[dfrating.rating <= 2]   #Делим на 3 группы по классу рейтинга. Это одна.
df_low
```

```python
df_medium = dfrating[(dfrating.rating > 2) & (dfrating.rating <= 4)]
df_medium
```

```python
df_high = dfrating[(dfrating.rating > 4) & (dfrating.rating <= 5)]
df_high
```

```python
df_high.rating.mean()   #Находим среднее для каждой группы
```

```python
df_medium.rating.mean()
```

```python
df_low.rating.mean()
```

```python
classification = dfrating.groupby('movieId').agg(['mean'])['rating'].head()  #Считаем среднее значение для каждого фильма.
classification.head()
```

```python
def class_(mean):   #Присваиваем категорию в зависимости от значения.
    if mean <= 2.0:
        return 'Low'
    elif mean <= 4.0 and mean > 2.0:
        return 'Medium'
    elif mean > 4.0 and mean <= 5.0:
        return 'High'
    
class_(4.3)
```

```python
dfrating['class_'] = dfrating['rating'].apply(class_)
dfrating.head(20)
```

Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.

```python
import pandas as pd
```

```python
dfkeywords = pd.read_csv('keywords.csv')
dfkeywords.head(30)  #выведем побольше строк и посмотрим содержание
```

```python
dfkeywords.info()       #Интересно, какой объем инфы в датасете
```

```python
keyword = 'одноклассники социальная сеть'
keyword.split(' ')           #Разбивка на слова
```

```python
keyword_split = lambda keyword: keyword.split(' ')   #Пишем лямбда-функцию
```

```python
dfkeywords['keyword_split'] = dfkeywords['keyword'].apply(lambda keyword: keyword.split(' '))
dfkeywords.head(10)                 #Записываем результат в отдельный столбец
```

```python
def region(line):
    if ('москва' in line['keyword']) or ('тула' in line['keyword']) or ('ярославль' in line['keyword']):
        return 'Центр'     
    elif ('петербург' in line['keyword']) or ('псков' in line['keyword']) or ('мурманск' in line['keyword']):
        return 'Северо-Запад'
    elif ('владивосток' in line['keyword']) or ('сахалин' in line['keyword']) or ('хабаровск' in line['keyword']):
        return 'Дальний Восток'
    return 'undefined'
```

```python
dfkeywords['region'] = dfkeywords.apply(region, axis=1)
dfkeywords.head()      #Вроде работает) 
```

```python
dfkeywords.region.value_counts()        #Проверим, сколько вхождений в разбивке по регионам.
```
