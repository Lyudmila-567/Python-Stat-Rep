# Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
q1 = filtr_columns['pulse'].quantile(0.25)
q3 = filtr_columns['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = filtr_columns[filtr_columns['pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('Age')
remove_outliers
```

```python
pd.concat([filtr_columns, remove_outliers]).drop_duplicates(keep=False)
```

Найдено 5 выбросов. 
184 - с большей вероятностью, неправильный ввод данных - уточнить у замерщика. Такое значение едва ли реально, в лит-ре встречается максимальный пульс 175, та и то во время бега. 
Остальные значения могут быть причиной:
1) неправильного измерения (во время бега) - взять методику измерений, расспросить замерщиков, узнать, тренируют ли лошадей до полугода
2) ошибкой ввода данных
3) маячком большой патологии. Мы видим, что все они подвергались хирургическому вмешательству и умерли все из них, кроме одной, которая не подвергалась операции (тип вмешательства - мочевой пузырь). Учитывая, что все особи относятся к молодняку, у которого средний пульс 90-120, то пульс 150 для патологий вполне допустим.

Таким образом, сначала соберем сведения и узнаем причины выбросов, а если такой возможности нет, то стоит заменить выбросы на 150 как максимальный для жеребцов. Потому что для патологий пульс 150 более чем приемлем в состоянии покоя.
